**Task - 3 - Outlier Detection:**

**Tasks:**
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.



```python
print("South Asia")
print(dsa[['Country name', 'score', 'Log GDP per capita']])

print("\nDescriptive Statistics for 'Score' and 'Log GDP per capita':")
print(dsa[['score', 'Log GDP per capita']].describe())
```

2. Define outliers using the 1.5 * IQR rule.

```python
def detect_outliers_IQR(column):
    Q1, Q3 = dsa[column].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower_bound, upper_bound = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    return (dsa[column] < lower_bound) | (dsa[column] > upper_bound)
dsa['Outlier_Score'] = detect_outliers_IQR('score')
dsa['Outlier_GDP'] = detect_outliers_IQR('Log GDP per capita')
dsa['Outlier'] = dsa['Outlier_Score'] | dsa['Outlier_GDP']
outliers = dsa[dsa['Outlier']]
print("\nOutlier Countries Based on Score and GDP per Capita:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.


```python
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=dsa, x='Log GDP per capita', y='score',
                               hue='Outlier', palette={True: 'green', False: 'red'}, s=100)

handles, labels = scatter_plot.get_legend_handles_labels()
scatter_plot.legend(handles=[handles[0], handles[1]], labels=['Normal', 'Outlier'], title='Indicator')

plt.title('Outlier Detection: GDP per Capita vs Score in South Asia')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.show()
```

These outliers can skew regional averages and impact policy decisions.

```python
dsa['Region'] = 'South Asia'
dme['Region'] = 'Middle East'

combined_df = pd.concat([dsa[['Country name', 'score', 'Region']],
                         dme[['Country name', 'score', 'Region']].rename(columns={'Score': 'score'})],
                        ignore_index=True)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='score', data=combined_df, palette='Set2')

plt.title('comparing the distribution of Score between South Asia and the Middle East.')
plt.xlabel('Region')
plt.ylabel('Happiness Score')

plt.show()
```

**Key Differences:**

Shape: South Asia's happiness scores may have more variation compared to the Middle East, suggesting more diverse experiences of well-being.

Median: The Middle East may have a higher median happiness score, indicating a generally higher average happiness compared to South Asia.

Outliers: Both regions may have outliers, but South Asia might show more extreme values, reflecting the regional disparities in happiness levels.

Interquartile Range (IQR): South Asia may show a wider IQR, indicating more variability, while the Middle East may have a more consistent experience of happiness among its countries.
